```python
import pandas as pd
```

## Primary data type in Pandas are:
### Series: 1d array which is size immutable
### Data Frame: 2d array

```python
s = pd.Series([10, 20, 30, 40, 50, 60])#the data are passed as a list to the Series function 
print(s)
a = pd.Series([1, 3, 4, 5, 6, 7,"apple"]) #the data are passed as a list to the Series function but it has mixed data types which makes the entire series of object data type
print(a)

# check values
print(s.values)

# check index
print(s.index)
```

## Giving a name to a column

```python
s.name = "Calories" #naming the series
print(s.name)
```

## Indexing :

```python
print(s[:3]) #slicing the series to get first three elements   

# Location based indexing
print(s.iloc[2]) #accessing the third element using iloc
print(s.iloc[[1, 3, 5]]) #accessing multiple elements using iloc    
print(s.iloc[1:5]) #accessing a range of elements using iloc
```

## changing the index names in pandas:

```python
index = [ 'apple', 'banana', 'orange', 'grape', 'kiwi', 'mango']
s.index = index  # assigning custom index to the series
#print(s)

##
s['orange']  # accessing element using custom index

#s.iloc['orange']  # this will raise an error because iloc only accepts integer location based indexing
#s.loc['orange']  # accessing element using custom index with loc


```

## label based indexing:

```python
s.loc['apple']
# In label based indexing both the start and end labels are included unlike in position based indexing where the end index is excluded
s.loc['orange':'mango']  # accessing a range of elements using loc with custom index
s.loc[['banana', 'kiwi', 'mango']]  # accessing multiple elements using loc with custom index
```

## creating serries using dictionary:

```python
fruit_protein = {
    'apple': 0.3,   
    'banana': 1.3,
    'orange': 1.0,
    'grape': 0.6,
    'kiwi': 1.1,
    'mango': 0.8
}

s2 = pd.Series(fruit_protein, name = "fruit_protein")     # creating a series from a dictionary
print(s2)
```

## Conditional Selection

```python
s2> 1.0  # filtering elements based on condition
```

```python
s2[s2 > 1.0]  # getting elements that satisfy the condition
```

## Logical operators:


```python
s2[(s2 > 1.0) & (s2 < 1.3)]  # getting elements that satisfy the condition
```

## modifying the series:

```python
s2['banana'] = 2.8  # updating value for a specific index
s2
```
