```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("shape of the matrix: \n", sales_data.shape)
print("sales data of first three restaraunt: \n",sales_data[0:3])
print("sales data of first three restaraunt without resturant id: \n",sales_data[:,1:])


```

## Total sales per year

```python
print("total sales of 2021: \n",np.sum(sales_data[:,1]))
print("total sale of 2022 \n:",np.sum(sales_data[:,1],axis=0)) # axis=0 means column wise operation
print("min sales of each resturant :\n", np.min(sales_data[:,1:],axis=1)) # axis=1 means row wise operation

## Max sales of the year :
print("max sale of the year :\n",np.max(sales_data[:,1:],axis=0))

## Average sales per year:
print("Average sales per year:\n",np.mean(sales_data[:,1:],axis=0))

## Average sales per resturant:
print("Average sales per year:\n",np.mean(sales_data[:,1:],axis=1))

## cumulative sales per resturant:
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print("cumulative sales per year:\n",cumsum)

## mean of cumulative sum:
print("Mean of the cumulative sum:\n",np.mean(cumsum,axis=0))

```

## Matplot to show the data in a graph

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales per year:")
plt.xlabel("Sales")
plt.ylabel("Years")
plt.grid(True)
plt.show()
```

## Vector operations

```python
vector1 = np.array([1,2,3])
vector2 = np.array([4,5,6])
print("vector addition:\n",vector1+vector2)
print("vector Multiplication:\n",vector1*vector2)

print("dot product",np.dot(vector1,vector2))
```
