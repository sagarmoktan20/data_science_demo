## Data Frame

```python
data = {
    "Name" : ["Alice_fernandes", "Bob_lama", "Charlie_chor", "David_becham", "Bob_lama"],
    "Age" : [24, 27, np.nan, 32,27],
    "Department" : ["HR", "Engineering", "Marketing", "Finance", "Engineering"],
    "salary" : [50000, 60000, 55000, np.nan, 60000]
}
print("\nData Dictionary:")
print(data)
```

```python
print("\nDataFrame from Dictionary:")
df = pd.DataFrame(data)
print(df)
```

# Data Celaning:

```python
## check for null values in the DataFrame
df.isnull()  # Check for null values in the DataFrame
df.isnull().sum()  # Count of null values in each column    
```

```python
## first way to drop rows with null values i.e if any null value is present in a row, that row will be dropped
tempdf = df.dropna(how='any')  # Drop rows with any null values, default is 'any' meaning if any null value is present in a row, that row will be dropped
print("\nDataFrame after dropping rows with null values:")  
print(tempdf) # Original DataFrame remains unchanged unless inplace=True is specified
```

```python
df.dropna(how='all') # Drop rows where all values are null i.e if all values in a row are null, that row will be dropped else it will be retained
```

```python
## second way of data cleaning is to fill the null values with some value
df.fillna(0000)  # Fill all null values with 0000 i.e replace null values with 0000 avoiding deletion of rows

df['Age'].fillna(df['Age'].mean())  # Fill null values in 'Age' column with the mean of that column, leaving other columns unchanged

df['salary'].fillna(df["salary"].median())  # Fill null values in 'salary' column with the median of that column, leaving other columns unchanged

```

### forward and backword fill in data cleaning

```python
df['Age'].fillna(method='ffill')  # Forward fill to propagate last valid observation forward to next valid  
df['Age'].fillna(method='bfill')  # Backward fill to propagate next valid observation backward to previous valid
```

### Replace a specific name in column

```python
df['Name'].replace('Alice_fernandes', 'shisir')  # Replace specific value in a column, here 'Alice' is replaced with 'shisir' in 'Name' column, it is not replaced in original DataFrame unless inplace=True is specified
df['Name'] = df['Name'].replace('Alice_fernandes', 'shisir')  # Replace specific value in a column permanently, here 'Alice' is replaced with 'shisir' in 'Name' column permanently because of assignment to the same column
print("\nDataFrame after replacing 'Alice_fernandes' with 'shisir' in 'Name' column:") 
print(df)
```

### Dealing with Duplicate data:

```python
df_dup = df[df.duplicated(keep = "first")]  # Check for duplicate rows in the DataFrame
df_dup
```

```python
df.drop_duplicates()  # Drop duplicate rows from the DataFrame, keeping the first occurrence by default, it does not modify the original DataFrame unless inplace=True is specified
```

## Dealing with edge or invalid cases:


```python
# Lambda functions and apply method can also be used for data cleaning and transformation but not shown here
df['promoted_salary'] = df['promoted_salary'].apply(lambda x: x * 1.1 if x < 120000 else x) # Increase 'promoted_salary' by 10% using apply and lambda function
print("\nDataFrame after increasing 'promoted_salary' by 10% for values less than 120000:")
print(df)
```

## spliting the first and last name:

```python
df['first', 'last'] = df['Name'].str.split('_')  # Split 'Name' column into 'first' and 'last' columns based on underscore delimiter
print("\nDataFrame after splitting 'Name' column into 'first' and 'last' columns:")
print(df)
```

## More on Lambda and Apply:

```python
def multiply_by_2(x):
    return x * 2    

df['Age'] = df['Age'].apply(multiply_by_2)  # Using apply method with a custom function to double the values in 'Age' column, here the original 'Age' column is updated with doubled values
print("\nDataFrame after doubling the values in 'Age' column:")
print(df)
```

```python
df['Age'] = df['Age'].apply(lambda x: x/2 if x>40 else x)
df["Age"]
```

## Joins and Merges
