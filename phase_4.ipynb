```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2  = np.random.rand(2,3)
arr3 = np.zeros((2,3))


np.save('arr1.npy',arr1)
np.save("arr2.npy",arr2)
np.save("arr3.npy",arr3)
```

```python
loaded_array1 = np.load('arr1.npy')
print(loaded_array1)
```

```python
try:
    logo = np.load("numpy-logo.npy")

    ##Display:
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo:")
    plt.grid(False)
    
    dark_logo = 1 - logo/255.0  
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Dark Numpy logo:")
    plt.grid(False)


except FileNotFoundError:
    print("Numpy logo not found!")
```
